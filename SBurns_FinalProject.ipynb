```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import os
%matplotlib inline
from matplotlib import pyplot as plt
import csv
import glob
```

```python
data = np.genfromtxt('CA_January2017.csv', skip_header = 20, skip_footer = 1, delimiter = ",")
print(data.shape)
ySize = np.size(data[:,15])
x = np.linspace(0,24,ySize)
y = data[:,15]
plt.plot(x,y, color = "B", linewidth = 0.5)
plt.title('Time vs. Temperature (ITS-90)')
plt.xlabel('Time [h]')
plt.ylabel('Temperature [C]')

total = 0

for i in y:
    #print(i)
    #print(total)
    total = i + total
avg = total/ySize
print(avg)   
```
